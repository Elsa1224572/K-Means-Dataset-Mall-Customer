1. import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import silhouette_score
```

3. load data

```python
df = pd.read_csv("Mall_Customers.csv")
df.head()
```

4. preprocessing

```python
le = LabelEncoder()
df['Genre'] = le.fit_transform(df['Genre'])  # Male=1, Female=0
```

5. memilih fitur untuk clustering

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

6. metode elbow untuk menentukan jumlah cluster

```python
inertia = []
K_range = range(1, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8,5))
plt.plot(K_range, inertia, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.grid()
plt.show()
```

7. fit K-Means dengan k=5

```python
kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
```

8. visualisasi clustering

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', data=df, palette='Set2')
plt.title("Customer Segmentation")
plt.grid()
plt.show()
```

9. Silhouette score

```python
score = silhouette_score(X, df['Cluster'])
print("Silhouette Score:", round(score, 3))
```

10. visualisasi hasil klasterisasi

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=df['Annual Income (k$)'],
    y=df['Spending Score (1-100)'],
    hue=df['Cluster'],
    palette='Set2',
    s=100
)
plt.title("Hasil Klasterisasi KMeans (k=5)")
plt.xlabel("Pendapatan Tahunan (k$)")
plt.ylabel("Spending Score (1100)")
plt.legend(title="Cluster")
plt.grid(True)
plt.show()
```

11. evaluasi

```python
from sklearn.metrics import silhouette_score

# Hitung Silhouette Score
score = silhouette_score(X, df['Cluster'])
print(f"Silhouette Score: {score:.4f}")
```
